```python
# for changing data types of extra curricular activities 
from sklearn.preprocessing import LabelEncoder,StandardScaler
```

```python
le=LabelEncoder()
df["Extracurricular Activities"]=le.fit_transform(df["Extracurricular Activities"])
```

```python
df
```

```python
#y=m1x1+m2x2+m3x3+m4x4+m5x5+c because we have 5 features 

x=df[["Hours Studied","Previous Scores","Extracurricular Activities","Sleep Hours","Sample Question Papers Practiced"]]
x
```

```python
y=df["Performance Index"]
y
```

```python
#splitting the data into a training data and testing data 

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2) # automatically training size 80%

x_train
```

```python
# to standardizing the data  we are using z scale= x-average/standrad deviation  , with standard scaler library
# optional and it is easy to read the data for the model 
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_train_scaled

```

```python
x_test_scaled=scaler.fit_transform(x_test)
x_test_scaled
```

```python
from sklearn.linear_model import LinearRegression
import pickle  # to store model in a physical file 
from sklearn.metrics import mean_squared_error,median_absolute_error, r2_score
```

```python
# training the model

model=LinearRegression()
model.fit(x_train_scaled,y_train)
```

```python
# testing the model

model.predict(x_test_scaled) # this x_test is nothing but predicted y
```

```python
x_test_scaled
```

```python
# we can also check for single record 

model.predict([[-1.57028968,  1.21427247,  1.05659932,  0.88175854,  0.167927  ]])

# we need to  find the metrics how accurate our model is but as of now we are good 
```

```python
y_pred=model.predict(x_test_scaled)  # this x_test is nothing but predicted y
y_pred
```

```python
mean_squared_error(y_test,y_pred)  # y test is actual y , y_pred is predicted y from x_test 
```

```python
r2_score(y_test,y_pred)
```

```python
import pickle
with open("Student-Performance.pkl",'wb') as file :
    pickle.dump((model,scaler,le),file)
```
